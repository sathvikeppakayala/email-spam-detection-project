```python
import pandas as pd
import numpy as np
import streamlit as st
```

```python
data = pd.read_csv("spam.csv", encoding = "latin-1")
data
```

```python
#getting columns of the Data
data.columns
```

```python
data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
```

```python
# Mapping Class and Message Columns
data['class'] = data['class'].apply(lambda x: 1 if x == "spam" else 0)
```

```python
data.head()
```

```python
# Importing Machine learning packages usable for Our classification
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
x = data['message']
y = data['class']
```

```python
print(x.shape)
print(y.shape)
```

```python
# Checking for NULL values if present
data.isnull().sum()
```

```python
cvect = CountVectorizer()
```

```python
x_vect = cvect.fit_transform(x)
```

```python
x_train, x_test,y_train, y_test = train_test_split(x_vect, y, test_size=0.50)
```

```python
print(x_train.shape)
```

```python
x_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
msg = "YOUR LAST CHANCE HURRYY UP Black Friday Timed Carnival Sale, Up to 50% Off!"
df = [msg]
vectorizer = cvect.transform(df).toarray()
prediction = model.predict(vectorizer)
```
